```python
url='https://en.wikipedia.org/wiki/S%26P_500'
response=requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
table = soup.find('table', class_= 'wikitable')
print(table)
```

```python
# prompt: i would like to extract data from this table into csv

import pandas as pd
import requests
from bs4 import BeautifulSoup

# Assuming the previous code is executed and 'table' variable is defined

# Extract table rows
rows = table.find_all('tr')

# Prepare data for DataFrame
data = []
for row in rows:
    cols = row.find_all(['th', 'td'])
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Create DataFrame
df = pd.DataFrame(data)

# Export to CSV
df.to_csv('returns_data.csv', index=False, header=False) # Set header to False if no header is desired.
```
